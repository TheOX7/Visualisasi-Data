# Pandas 27 | Fungsi agregasi first dan last

---





## Persiapan Data Frame

```python
d = {
    'dokter' : ['Budi', 'Wati', 'Iwan', 'Budi', 'Budi', 'Wati'],
    'pasien' : ['Abdul', 'Rahmat', 'Asep', 'Joko', 'Wiwin', 'Lisa']
}

df = pd.DataFrame(d)
df
```


## Fungsi agregasi **first()** dan **last()**




```python
df.groupby('dokter')['pasien'].count().to_frame()
```

```python
df.groupby('dokter')['pasien'].first().to_frame( )
```

```python
df.groupby('dokter')['pasien'].last().to_frame( )
```

# Pandas 28 | Mengenal explode dan implode List pada Pandas Data Frame

---





## Persiapan Data Frame

```python
d = {
    'Team' : ['DC', 'Marvel'],
    'Heroes' : [['Batman', 'Superman', 'Wonder Woman', 'Aquaman', 'Green Lanters', 'Shazam'], 
                ['Iron Man', 'Captain America', 'Ant-Man', 'Black Panther', 'Captain Marvel']]
}

df = pd.DataFrame(d)
df
```


## Explode()




```python
df1 = df.explode('Heroes')
df1
```


## Implode()




```python
d = {'Team' : ['DC', 'Marvel']}
df2 = pd.DataFrame(d)
df2
```

```python
df2['Imploded'] = df1.groupby(df1.index)['Heroes'].agg(list)
df2
```
