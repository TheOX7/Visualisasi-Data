# Pandas 50 | Membuat kolom baru dengan for loop dan f-string

---





## Persiapan Data Frame

```python
d = {
    'penjual' : ['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'],
    'barang' : ['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']
}

df = pd.DataFrame(d)
df
```


## Membuat kolom baru dengan **for loop** dan **f-string**

```python
cols = ['penjual', 'barang']

for col in cols :
  df[f'count_tiap{col}'] = df.groupby(col).cumcount()+1

df
```

# Pandas 51 | Seleksi Baris dengan between

---





## Persiapan Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size = (n_rows, n_cols)), columns = cols)
df
```


## Seleksi baris dengan **between**

```python
df[df['A'].between(2, 5)] # Dalam keadaan inclusive, jadi 2 & 5 termasuk
```

```python
df[df['A'].between(2, 5, inclusive=False)] # Dalam keadaan exclusive, jadi 2 & 5 tidak termasuk
```

# Pandas 52 | Transformasi kolom menjadi baris

---





## Persiapan Data Frame

```python
d = {
    'kode_area' : [123, 456, 789, 321],
     'pabrik' : [4, 2, 5, 6],
     'gudang' : [2, 4, 7, 3],
     'toko' : [64, 32, 15, 24]
}

df = pd.DataFrame(d)
df
```


## Transformasi kolom menjadi baris

```python
df = df.melt(
    id_vars='kode_area',
    var_name='jenis_bangunan',
    value_name='jumlah'
)

df
```
