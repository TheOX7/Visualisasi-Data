# Pandas 41 | Agregasi pada banyak kolom Data Frame dengan agg

---





## Persiapan Data Frame

```python
df = pd.read_csv('/content/titanic_full.csv')
df.head()
```


## Menerapkan agregasi pada sejumlah kolom dengan **agg()**



```python
df.groupby('pclass').agg({
    'pclass' : 'count',
    'age' : ['mean', 'max'],
    'survived' : 'mean'
})
```

```python
df.groupby('pclass').agg(
    n_pass = ('pclass', 'count'),
    avg_age = ('age', 'mean'),
    max_age = ('age', 'max'),
    survival_rate = ('survived', 'mean'),
)
```

# Pandas 43 | Menangani whitespace pada Pandas Data Frame

---





## Persiapan Data Frame

```python
data = {
    'nim' : ['10', '11', '12', '13', ' '],
    'name': ['adi', ' ', 'tejo', ' ', 'bejo']
}

df = pd.DataFrame(data)
df
```


## Menangani whitespace pada Dataframe

```python
df
```

```python
df.info()
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

```python
df.info()
```
