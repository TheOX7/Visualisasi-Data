# Pandas 09 | Membagi Data Frame menjadi dua secara acak 

---





## Persiapan Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('VXYWZ')

df = pd.DataFrame(np.random.randint(1, 20 , size=(n_rows, n_cols)), columns=cols)
df
```

## Membagi Data Frame menjadi dua secara acak berdasarkan proporsi tertentu





```python
# Untuk mendapat jumlah baris dan kolom --> Output (jmlh_baris, jmlh_kolom)
df.shape
```

```python
proporsi = 0.7 # 70% untuk df_1 dan 30% untuk df_2

# Fungsi frac --> Pada contoh ini, untuk mengambil baris secara acak dari data df sebanyak 70%
df_1 = df.sample(frac=proporsi)

# Lalu kita drop/hapus berdasarkan index/baris dari df_1
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}') # Output --> 70% baris yg diambil, jadi ada 7 baris
print(f'df_2 shape: {df_2.shape}') # Output --> 30$ baris yg diambil, jadi ada 3 baris
```

```python
# Hasil barisnya jadi acak, karena pake fungsi frac tadi
df_1
```

```python
df_2
```

## Membalik Urutan Baris dan Melakukan Penyesuaian Ulang **Index**
Tujuannya agar me-reverse baris namun indexnya dimulai dari angka terkecil dan hasilnya tetap sesuai dengan hasil reverse sebelumnya

```python
df.loc[::-1].reset_index(drop=True)
```

# Pandas 10 | Mengganti nama kolom pada Data Frame berdasarkan pola

---





## Persiapan Data Frame

```python
df = pd.read_csv('/content/titanicfull.csv')
df.columns = ['Pclass', 'Survival status', 'full Name', 'Sex  ', '  Age',
              'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']

df_backup = df.copy(deep=True)
df.head()
```

## Menggunakan lowercase untuk nama kolom dan mengganti spasi dengan ( **_** )





```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

## Memangkas kelebihan spasi pada nama kolom

```python
df = df_backup.copy(deep=True)

# Fungsi strip() --> Untuk menghapus spasi yang ada di sebelum & setelah kata
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.head()
```

# Pandas 11 | Seleksi kolom dan baris pada Data Frame menggunakan loc

---





## Persiapan Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('VXYWZ')

df = pd.DataFrame(np.random.randint(1, 20 , size=(n_rows, n_cols)), columns=cols)
df
```

## Seleksi kolom dan baris menggunakan **loc**





```python
df.loc[[0,3,4], ['X', 'Z']]
```

## Seleksi baris dengan kondisi

```python
df.loc[df['V']>10, ['V', 'X', 'Y', 'W', 'Z']]
```

## Slicing Data Frame dengan **loc**

```python
df.loc[0:4, 'V':'Z']
```

# Pandas 12 | Membentuk kolom bertipe datetime dari beberapa kolom lain

---





## Persiapan Data Frame

```python
data = {'day':[1, 2, 10, 25, 12],
        'month':[1, 2, 4, 5, 6],
        'year':[2000, 2001, 2010, 2015, 2020]
        }

df = pd.DataFrame(data)
df
```

**bold text**## Seleksi kolom dan baris menggunakan **loc**





```python
df.loc[[0,3,4], ['X', 'Z']]
```

## Membentuk kolom bertipe **datetime**

```python
df['penanggalan'] = pd.to_datetime(df[['day', 'month', 'year']])
df
```
