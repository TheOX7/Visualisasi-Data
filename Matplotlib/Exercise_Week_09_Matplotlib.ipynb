```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan temporary style

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
# Supaya temporary/sementara plottingannya pakai klausa 'with'
with plt.style.context('dark_background'):
  plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'b-o')

plt.show()
```

```python
# Jadi kembali ke semula
plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```
