## Animasi pada Matplotlib

```python
fig = plt.figure()
plt.xlim((0,4))
plt.ylim((-2,2))

line = plt.plot([])[0]

def animation(i):
  x = np.linspace(0, 4, 1000)
  y = np.sin(2 * np.pi * (x - 0.01 * i))
  line.set_data(x,y)
  return line

animate = FuncAnimation(fig, animation, frames=500, interval=50)

video = animate.to_html5_video()
html = display.HTML(video)
display.display(html)

# plt.close()
```
