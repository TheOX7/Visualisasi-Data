# Pandas 21 | Agregasi pada Pandas Data Frame dengan agg

---





## Persiapan Data Frame

```python
df = pd.read_csv('/content/Iris.csv')
df
```


## Mengenal **grouby()** dan fungsi agregasi





```python
df.groupby('Species')['PetalLengthCm'].count().to_frame()
```

```python
df.groupby('Species')['PetalLengthCm'].mean().to_frame()
```

```python
df.groupby('Species')['PetalLengthCm'].median().to_frame()
```


## Agregasi dengan **agg()**





```python
df.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median'])
```


## Agregasi dengan **describe()**





```python
df.groupby('Species')['PetalLengthCm'].describe()
```

---

# Pandas 24 | UTC dan konversi zona waktu 

---





## Persiapan Series

```python
s = pd.Series(range(1591683521, 1592201921, 3600)) 
# (1591683521, 1592201921) --> Unix Epoch Time
# 3600 --> Jumlah detik dlm 1 jam

s = pd.to_datetime(s, unit='s') # unit = 's' --> dlm satuan detik/second
s.head()
```


## Pengaturan Zona Waktu (Time Zone)



```python
s = s.dt.tz_localize('UTC')
s.head()
```

```python
# Diconvert dari UTC menjadi jam/pukul pada Jakarta
s = s.dt.tz_convert('Asia/Jakarta')
s.head()
```

```python
# Diconvert dari Asia/Jakarta menjadi jam/pukul pada Hobart
s = s.dt.tz_convert('Australia/Hobart')
s.head()
```
